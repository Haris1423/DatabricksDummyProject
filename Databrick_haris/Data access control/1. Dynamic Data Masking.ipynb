## CREATE THE MASK FUNCTION


- CREATE NEW GROUPS LIKE DEVELOPERS, ADMIN

```python
%sql
CREATE OR REPLACE FUNCTION databricksharis.silver.dynamic_mask(p_email STRING)
RETURNS STRING
LANGUAGE SQL
RETURN CASE WHEN is_account_group_member('developers') then p_email else '*****' END
```

## APPLY MASK 

```python
%sql
ALTER TABLE databricksharis.silver.customers_enr 
ALTER COLUMN email SET MASK databricksharis.silver.dynamic_mask
```
