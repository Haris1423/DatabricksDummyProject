```sql
---blend of parameters and qquery snippets
---- top 3 best performing products for our organisation

with joined_query as (
select 
sum(s.total_amount) total_amount,
p.product_name
from sales_enr s
left join products_enr p on s.product_id = p.product_id
group by p.product_name
)

--- let's say we can create a parameter for top 3 or top 5 
,
ranked_query as (
select *,dense_rank() over( order by total_amount desc) as ranking from joined_query
)

--- he has to write the top of customers
select product_name, total_amount from ranked_query where ranking <=  :Top

---We can schedule the query by clicking on the calender button


```
