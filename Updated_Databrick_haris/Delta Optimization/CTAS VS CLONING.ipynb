## **CTAS** COPY TABLE AS STAYEMENT
It'll create Table on existing Table 

![image_1773507062732.png](./image_1773507062732.png "image_1773507062732.png")

```python
%sql
CREATE TABLE databricksharis.silver.ctas AS
SELECT * FROM databricksharis.silver.sales_enr


--- DOn't wprrry about the history only data 
```

```python
%sql
SELECT * FROM databricksharis.silver.CTAS
```

### **DEEP `CLONE`** : 
![image_1773589859686.png](./image_1773589859686.png "image_1773589859686.png")
- IN THIS IT'LL NOT ONLY CLONE THE DATA BUT THE DELTA LOG AS WELL, WE CAN GO BACK IN TIME ALSO

```python
%sql
CREATE OR REPLACE TABLE databricksharis.silver.deepclone
CLONE databricksharis.silver.sales_enr
```

```python
%sql
SELECT * FROM databricksharis.silver.deepclone
```

WITH CLONING , WE CLONING CURRENT STATE OF THE TABLE

### brings the current snapshot of the table And also difference between Deep Clone and CTAS in CTAS we can make table with filter with any state

```python
%sql
---it version Starts from 0 
DESCRIBE HISTORY databricksharis.silver.deepclone

```

## **SHALLOW CLONE**
- LIGHT WEIGHT CLONE, DATA WILL NOT BE CLONE ONLY DELTA LOG WILL BE CLONED , WE'LL NOT LIFT AND SHIFT THE DATA, DATA WILL COME FROM ORIGINAL SOURCE. DISADVANTAGE IS LIKE IF CHANGING IN SOURCE, IT'LL ALSO BE REFLECTING IN SAME

- highly Dependent on source, let's say some partition are break in source, it will break as it's delta log is independent and depends on the source.

```python
%sql
CREATE TABLE databricksharis.silver.shallowclone
SHALLOW CLONE databricksharis.silver.sales_enr
```
