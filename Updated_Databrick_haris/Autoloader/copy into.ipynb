### COPY INTO IS SUITABLE FOR THE LOW AND MED RANGE DATA 

Copy files really works with parquet files as csv don't carry the schema so for this we use parquet files.
And for Copy into we need Table also

Create a directory copy_into and upload files

```python
%sql 
----parquet files carry real schema instead of c0,c1 and all in footer of the files

SELECT * FROM parquet.`/Volumes/databricksharis/bronze/auto_vol/copy_into/part-00000-tid-1800520145580676501-2fb00642-a896-4599-9957-2c7260f04aa7-132-1.c000.snappy.parquet`
```

Create Schma after reading the file , not just put random things

```python
%sql
CREATE TABLE databricksharis.bronze.copy_into (
  order_id int,
  customer_id int,
  order_date date,
  order_amount DOUBLE,
  order_status string
)
```

Why we defined two times here, in Autoloader we use add new columns, format_options : if we recieve any mismatch, and copy options is for tables and merge columns in tables similar to mergeSchema= True as we re run the query and refersh in Autoloader

```python
%sql
COPY INTO databricksharis.bronze.copy_into
FROM "/Volumes/databricksharis/bronze/auto_vol/copy_into/"
FILEFORMAT = PARQUET
FORMAT_OPTIONS ("mergeSchema" = "true")
COPY_OPTIONS ("mergeSchema" = "true")
```

```python
%sql
SELECT * FROM databricksharis.bronze.copy_into
```
