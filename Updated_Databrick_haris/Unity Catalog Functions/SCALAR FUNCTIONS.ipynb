## SQL SCALAR FUNCTIONS

Normal pyspark functions are session scoped , but functions in unity catalog is not session scoped , we can use it anywhere in Unity catalog

TAKES SCALAR VALUES AND RETURN THE SCALAR VALUES

![image_1773377403679.png](./image_1773377403679.png "image_1773377403679.png")

```python
%sql
CREATE OR REPLACE FUNCTION databricksharis.bronze.sql_scalar_function(p_par int)
RETURNS INT 
LANGUAGE  SQL
RETURN p_par *10 ;
```

```python
%sql
select total_amount,databricksharis.bronze.sql_scalar_function(total_amount) from databricksharis.silver.sales_enr
```

## PYTHON FUNCTION

```python
%sql
CREATE OR REPLACE FUNCTION databricksharis.bronze.python_scalar_function(p_par INT)
RETURNS INT
LANGUAGE PYTHON 
AS
$$
    return p_par *10
$$
```

```python
%sql
select total_amount,databricksharis.bronze.python_scalar_function(total_amount) from databricksharis.silver.sales_enr
```
