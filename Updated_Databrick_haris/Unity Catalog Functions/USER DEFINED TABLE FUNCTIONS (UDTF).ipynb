## USER DEFINED TABLE FUNCTIONS (UDTF)

IF WE PASS THE SCALAR VALUES, IT RETURNS TABLE DATA
- FLOW
- PARAMETER ---> TABLE ----> UDTF----> TABLE'S DATA


IT WILL WRITE SOME LOGIC CARRY SOMETHING AS SELECT STATEMENT, AND WITH IN THE STATEMENT IT WILL CARRY THE KIND OF PARAMETERS. AND WHEN IT WILL GET THE PARAMETER IT'LL GIVE THE RESULTS

```python
%sql
CREATE OR REPLACE FUNCTION databricksharis.bronze.UDTF(p_para INT)
RETURNS TABLE 
LANGUAGE SQL --BY DEFAULT IT'S SQL
RETURN
(

  SELECT * FROM databricksharis.silver.sales_enr where total_amount > p_para
)
```

you don't need to treat it as normal functions, we can't use this for columns as it returns table data

```python
%sql
SELECT * FROM databricksharis.bronze.UDTF(100) -- it will filter out the results
```
