### Let's say i'm the owner of East region

### **MAPPING TABLE**

- First We create Mapping Table and tell us Which user belongs to which Region

```python
%sql
CREATE TABLE databricksharis.silver.rls_mapping(
  user_id STRING,
  region STRING
)
```

```python
%sql
INSERT INTO databricksharis.silver.rls_mapping(user_id,region) VALUES('harisbhat14231@gmail.com','east'),('kishan.daxini@polestarllp.com','west'),('waseem.wani@polestarllp.com','north'),('neeraj.kumar@polestarllp.com','south')
```

```python
%sql
select * from databricksharis.silver.rls_mapping
```

```python
%sql
select * from databricksharis.silver.rls_mapping
where user_id = current_user()
and region = 'east'
```

###  will not get any result

```python
%sql
select * from databricksharis.silver.rls_mapping
where user_id = current_user()
and region = 'west'
```

## WE WILL USE BOLEAN

```python
%sql
SELECT EXISTS 
(
select * from databricksharis.silver.rls_mapping
where user_id = current_user()
and region = 'west'
)
```

### RLS FUNCTION

```python
%sql
CREATE OR REPLACE FUNCTION databricksharis.silver.rls_function(p_region STRING)
RETURNS boolean
RETURN
EXISTS 
(
select * from databricksharis.silver.rls_mapping
where user_id = current_user()
and region = lower(p_region)
)
```

### APPLY NOW FUNCTION ON COLUMN OF A TABLE

```python
%sql
ALTER TABLE databricksharis.silver.stores_enr
SET ROW FILTER databricksharis.silver.rls_function ON (region)
    
```

```python
%sql
SELECT * FROM databricksharis.silver.stores_enr
```
