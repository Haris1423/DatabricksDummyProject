### WHY DELTA SHARING?

We have metastore, and we have one workspace, and there is other attached workspace to this metastore, both can share their data without any protocal.

Let's say we have different Meta Stores and it has a catalog and schema, and these catalog cannot access the catalog or data from earlier metstaore, Here comes delta sharing.

We can say metastore per region
